```python
import pandas as pd
import numpy as np
import os
from google.colab import drive
drive.mount('/content/drive')

import os
path = os.getcwd()
```

```python
headlines = []
for filename in os.listdir(path+"/drive/MyDrive"):
    if "Articles" in filename:
        article_df = pd.read_csv(path + "/drive/MyDrive/" + filename)
        headlines.extend(list(article_df["headline"].values))
```

```python
len(headlines)
```

```python
headlines = [ h for h in headlines if h != "Unknown" ]
print("The number of headline is :", len(headlines))
```

```python
import string
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action='ignore', category=FutureWarning)

def clean_text(headline):
    text = "".join( word for word in headline if word not in string.punctuation ).lower()
    text = text.encode("utf8").decode("ascii", "ignore")
    return text

corpus = [ clean_text(headline) for headline in headlines ]
```

```python
vocab = []
for line in corpus:
    words = line.split()
    for word in words:
        vocab.append(word)

vocabulary = set(vocab)
```

```python
len(vocabulary)
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.utils import to_categorical

tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word2index = tokenizer.word_index
len(word2index)
```

```python
word2index
```

```python
dictionary = {}
rev_dictionary = {}
for word, idx in word2index.items():
    dictionary[word] = idx
    rev_dictionary[idx] = word
```

```python
rev_dictionary
```

```python
input_sequences = tokenizer.texts_to_sequences(corpus)
```

```python
input_sequences
```

```python
input_data = []
target = []
for line in input_sequences:
    for i in range(1, len(line)-1):
        input_data.append(line[:i])
        target.append(line[i])
```

```python
max(target)
```

```python
MAX_LEN = 0
for seq in input_data:
    if len(seq) > MAX_LEN:
        MAX_LEN = len(seq)
MAX_LEN
```

```python
input_data = pad_sequences(input_data, maxlen=MAX_LEN,
                           padding="post")
```

```python
input_data.shape
```

```python
target = to_categorical(target, num_classes=len(word2index))
```

```python
from keras.models import Sequential
from keras.layers import Embedding, GRU, Dense, Dropout, LSTM
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Embedding(input_dim=len(vocabulary), output_dim=100,
                    input_length=MAX_LEN))

model.add(LSTM(units=100))
model.add(Dropout(rate=0.1))

model.add(Dense(units=target.shape[1], activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer="adam", metrics=['accuracy'])
```

```python
model.fit(input_data, target, batch_size=10, epochs=10)
```

```python
new_sent = 'Trump'
seq = tokenizer.texts_to_sequences([new_sent])
new_input = pad_sequences(seq, maxlen=MAX_LEN,
                           padding="post")
print(new_input.shape)
result=np.argmax(model.predict(new_input, verbose=False))
rev_dictionary[result]
```
