#### Data Declaration

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabets)
```

#### Data Preparation

```python
int_to_char = dict(enumerate(alphabets))
```

```python
int_to_char
```

```python
char_to_int = dict((v,i) for i, v in enumerate(alphabets))
```

```python
char_to_int
```

```python
seq_length = 1
```

```python
x = []
y = []
for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in,'->',seq_out)
```

```python
x
```

```python
y
```

#### One-hot encoding for y

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
y_new
```

```python
import numpy as np
```

```python
np.array(x).shape
```

```python
x = np.reshape(x, (25,1,1))
```

```python
x.shape
```

```python
x = x / 25
```

```python
x
```

Build the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',
              metrics = ['accuracy'])
```

```python
model.fit(x, y_new, epochs=500, batch_size=1)
```

```python
model.evaluate(x, y_new, batch_size=1)
```

Predict on the new data

```python
new = 'V'
new = char_to_int[new]
new = np.reshape(new, (1,1,1))
new = new / 25
result = model.predict(new, verbose=False)
result = np.argmax(result)
print(int_to_char[result])
```

```python
seq_length = 3
```

```python
x = []
y = []
for i in range(len(alphabets) - seq_length):
  seq_in = alphabets[i:i + seq_length]
  seq_out = alphabets[i + seq_length]
  x.append([char_to_int[char] for char in seq_in])
  y.append(char_to_int[seq_out])
  print(seq_in,'->',seq_out)
```

```python
x = np.reshape(x, (len(x), seq_length, 1))
```

```python
x
```

```python
x = x / 25
```

```python
x.shape
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(3,1), activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation = 'softmax'))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(x, to_categorical(y), epochs=500, batch_size=1)
```

```python
model.evaluate(x, to_categorical(y) , batch_size=1)
```

```python
new = 'BCD'
val = np.array([[char_to_int[x]] for x in new])
val = val.reshape(1,3,1)
val = val / 25
result = model.predict(val)
result = np.argmax(result)
print(int_to_char[result])
```
